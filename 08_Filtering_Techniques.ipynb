```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image =cv2.imread(r"/content/img.jpg")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
# plt.axis('off')  # Turn off axis numbers
plt.show()
```

#### Box filters, filter 2D, Gaussian Blur, and Median Blur

### Box Filter

```python
box_filtered = cv2.boxFilter(image_rgb, -1, (9, 2))

# Display result
plt.imshow(box_filtered)
plt.title('Box Filter')
plt.axis('off')
plt.show()
```

### 2D Filter (Arbitrary Kernel)

```python
# Define a sharpening kernel
kernel = np.array([[0, -1, 0],[-1, 5, -1], [0, -1, 0]])

# Apply the filter using filter2D
filtered_image = cv2.filter2D(image_rgb, -1, kernel)

# Display result
plt.imshow(filtered_image)
plt.title('2D Filter (Sharpening)')
plt.axis('off')
plt.show()
```

### Gaussian Blur

```python
# Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(image_rgb, (5, 5), sigmaX=1)

# Display result
plt.imshow(gaussian_blur)
plt.title('Gaussian Blur')
plt.axis('off')
plt.show()
```

### Median Blur

```python
# Apply Median Blur
median_blur = cv2.medianBlur(image_rgb, 9)

# Display result
plt.imshow(median_blur)
plt.title('Median Blur')
plt.axis('off')
plt.show()
```
