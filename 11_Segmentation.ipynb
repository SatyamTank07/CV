#### Performing Segmentation

```python
image_gray = cv2.imread(r"/content/img.jpg",cv2.IMREAD_GRAYSCALE)
```

### Simple Thresholding

```python
def simple_threshold(image, threshold_value=127, max_value=255):
  _, thresholded_image = cv2.threshold(image, threshold_value, max_value, cv2.THRESH_BINARY)
  return thresholded_image

# Apply simple thresholding
simple_thresh_image = simple_threshold(image_gray, threshold_value=127)
plt.imshow(simple_thresh_image, cmap='gray')
plt.title('Simple Thresholding')
plt.axis('off')
plt.show()
```

### Otsu’s Binarization

```python
def otsu_threshold(image):
  _, otsu_thresh_image = cv2.threshold(image, 0, 255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
  return otsu_thresh_image

# Apply Otsu's thresholding
otsu_thresh_image = otsu_threshold(image_gray)
plt.imshow(otsu_thresh_image, cmap='gray')
plt.title("Otsu's Binarization")
plt.axis('off')
plt.show()
```

### Adaptive Thresholding

```python
def adaptive_threshold(image, method='mean', block_size=11, C=2):
  if method == 'mean':
      adaptive_thresh_image = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, block_size, C)
  elif method == 'gaussian':
      adaptive_thresh_image = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, block_size, C)
  return adaptive_thresh_image

# Apply adaptive mean thresholding
adaptive_mean_image = adaptive_threshold(image_gray, method='mean')
plt.imshow(adaptive_mean_image, cmap='gray')
plt.title('Adaptive Mean Thresholding')
plt.axis('off')
plt.show()
```

```python
# Apply adaptive Gaussian thresholding
adaptive_gaussian_image = adaptive_threshold(image_gray,method='gaussian')
plt.imshow(adaptive_gaussian_image, cmap='gray')
plt.title('Adaptive Gaussian Thresholding')
plt.axis('off')
plt.show()
```
