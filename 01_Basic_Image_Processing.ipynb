### Read an Image in OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image =cv2.imread(r"/content/img.jpg")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
# plt.axis('off')  # Turn off axis numbers
plt.show()
```

### Resize the image

```python
print("before : ", image_rgb.shape[:2])
resized_image = cv2.resize(image_rgb, (200, 300))
plt.imshow(resized_image)
plt.title('Resized Image')
plt.axis('off')
plt.show()
print("after : ", resized_image.shape[:2])
```

### Slicing a part of the image

```python
sliced_image = image_rgb[100:300, 150:700]
plt.imshow(sliced_image)
plt.title('Sliced Part of the Image')
plt.axis('off')
plt.show()
```

### Mask a part of image

```python
maskImg = image_rgb.copy()
for i in range(image_rgb.shape[0]):
    for j in range(image_rgb.shape[1]):
      if (i<100 and j <100) or (i<100 and j > 1180) or (i>620 and j < 100)  or (i>620 and j > 1180):
        maskImg[i][j]=[0,0,0]
plt.imshow(maskImg)
```
