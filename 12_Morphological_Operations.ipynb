```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image =cv2.imread(r"/content/img.jpg")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
# plt.axis('off')  # Turn off axis numbers
plt.show()
```

```python
image_gray = cv2.imread(r"/content/img.jpg",cv2.IMREAD_GRAYSCALE)
```

### Performing Morphological Operations 🎉🎉🎉

## Erosion

```python
def erosion(image, kernel_size=3):
  kernel = np.ones((kernel_size, kernel_size), np.uint8) # Define kernel size
  eroded_image = cv2.erode(image, kernel, iterations=1) # Perform erosion
  return eroded_image

# Apply erosion
eroded_image = erosion(image_gray)
plt.imshow(eroded_image, cmap='gray')
plt.title('Erosion')
plt.axis('off')
plt.show()
```

## Dilation


```python
def dilation(image, kernel_size=3):
  kernel = np.ones((kernel_size, kernel_size), np.uint8) # Define kernel size
  dilated_image = cv2.dilate(image, kernel, iterations=1) # Perform dilation
  return dilated_image

# Apply dilation
dilated_image = dilation(image_gray)
plt.imshow(dilated_image, cmap='gray')
plt.title('Dilation')
plt.axis('off')
plt.show()
```

## Opening

```python
def opening(image, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8) # Define kernel size
    opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel) #Perform opening
    return opened_image

# Apply opening
opened_image = opening(image_gray)
plt.imshow(opened_image, cmap='gray')
plt.title('Opening')
plt.axis('off')
plt.show()
```

## Closing

```python
def closing(image, kernel_size=3):
  kernel = np.ones((kernel_size, kernel_size), np.uint8) # Define kernel size
  closed_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel) # Perform closing
  return closed_image

# Apply closing
closed_image = closing(image_gray)
plt.imshow(closed_image, cmap='gray')
plt.title('Closing')
plt.axis('off')
plt.show()
```
